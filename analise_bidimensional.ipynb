```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando dados do Kaggle
path = "/content/drive/MyDrive/data_science_studies_material/insurance.csv"
data = pd.read_csv(path)

data.info()
data.describe()

# Selecionando as colunas desejadas para análise bidimensional
x = data["bmi"]
y = data["charges"]

# Criando a plotagem
plt.scatter(x, y)
plt.xlabel('BMI')
plt.ylabel('Custo do Seguro')
plt.title('Análise Bidimensional de Custo de Seguro com base no BMI')

# Salvando a plotagem como imagem
plt.savefig('analise_bidimensional_kaggle.png')

correlation = round(data['bmi'].corr(data['charges']), 2)
print(correlation)
```
